# Exponential Function - Negative Power
* A exponential function to a negative power gets smaller as x gets bigger
* So this function indicates that time going forward **lessens** the effect of the exponential function.
* A higher risk-free rate results in a faster decay - see below
* An exponential function to a positive power - does the opposite - this is compound interest

```python
from pylab import plt, mpl
plt.style.use('seaborn')
mpl.rcParams['font.family'] = 'serif'
%matplotlib inline
import numpy as np
def generate_risk_free_space(r):
    time = np.linspace(0, 20, 50)
    time = np.exp(-r * time)
    return time
```

```python
exp_space = generate_risk_free_space(0.25)
plt.plot(exp_space)
```

```python
exp_space = generate_risk_free_space(0.05)
plt.plot(exp_space)
```

```python
exp_space = generate_risk_free_space(-0.25)
plt.plot(exp_space)
```

# Ln(Price/Strike)
* A downwards sloping log that crosses x=0 at **Price**

```python
def ln_price_strike(price):
    strike_space = np.linspace(1, price+20, 100)
    strike_space = np.log(price/strike_space)
    return strike_space
```

```python
ln_price_hund = ln_price_strike(100)
plt.plot(ln_price_hund)
```
